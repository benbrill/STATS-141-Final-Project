```python
import pandas as pd
import numpy as np
```

```python
position_war = pd.read_csv('position_war.csv')
position_war
```

```python
pitcher_war = pd.read_csv('pitcher_war.csv')
pitcher_war
```

```python
position_minors = pd.read_csv("position_minors.csv")
position_major = pd.read_csv("position_majors.csv")
pitcher_majors = pd.read_csv("pitchers_majors.csv")
pitcher_minors = pd.read_csv("pitchers_minors.csv")
```

```python
position = pd.concat([position_minors, position_major]).sort_values(by=['name', 'gameDate'])
position
```

```python
pitchers = pd.concat([pitcher_majors, pitcher_minors]).sort_values(by=['name', 'gameDate'])
pitchers
```

```python
position["gameDate"] = pd.to_datetime(position["gameDate"])
position['season'] = pd.DatetimeIndex(position['gameDate']).year
position["days_between_game"] = position.groupby(["name", "season"])["gameDate"].transform(lambda X: X.diff())
position["days_between_game"] = position["days_between_game"].apply(lambda x: x.days - 1)
position.reset_index(inplace=True)
pitchers["gameDate"] = pd.to_datetime(pitchers["gameDate"])
pitchers['season'] = pd.DatetimeIndex(pitchers['gameDate']).year
pitchers["days_between_game"] = pitchers.groupby(["name", "season"])["gameDate"].transform(lambda X: X.diff())
pitchers["days_between_game"] = pitchers["days_between_game"].apply(lambda x: x.days - 1)
pitchers.reset_index(inplace=True)
```

```python
pitchers
```

```python
pitchers.sort_values("days_between_game", ascending=False)[~pitchers["league"].isin(["National League", "American League"])]
```

```python
import seaborn as sns
from sklearn.linear_model import LinearRegression

position_days_missed = position.groupby("name").apply(lambda X: X[~X["league"].isin(["National League", "American League"])][["days_between_game"]].max())
graph_df = position_war.merge(position_days_missed, left_on="Name", right_on="name", how="left")

graph_df["standard_WAR"] = (graph_df["WAR"] / graph_df["G"]) * 150
graph_df = graph_df[graph_df["G"] > 100]
# graph_df["days_between_game"] = graph_df["days_between_game"].fillna(0)
sns.scatterplot(data=graph_df, x="days_between_game", y="standard_WAR")
graph_df["days_between_game"] = graph_df["days_between_game"].fillna(0)

X = graph_df["days_between_game"].values.reshape(-1, 1)
y = graph_df["standard_WAR"].values.reshape(-1, 1)
reg = LinearRegression()
reg.fit(X, y)
reg.score(X, y)
```
