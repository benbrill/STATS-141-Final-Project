```python
position_minors = pd.read_csv("position_minors.csv")
position_major = pd.read_csv("position_majors.csv")
pitcher_majors = pd.read_csv("pitchers_majors.csv")
pitcher_minors = pd.read_csv("pitchers_minors.csv")
```

```python
position = pd.concat([position_minors, position_major]).sort_values(by=['name', 'gameDate'])
position
```

```python
pitchers = pd.concat([pitcher_majors, pitcher_minors]).sort_values(by=['name', 'gameDate'])
pitchers
```

```python
position["gameDate"] = pd.to_datetime(position["gameDate"])
position['season'] = pd.DatetimeIndex(position['gameDate']).year
position["days_between_game"] = position.groupby(["name", "season"])["gameDate"].transform(lambda X: X.diff())
position["days_between_game"] = position["days_between_game"].apply(lambda x: x.days - 1)
position[position["days_between_game"] < 0] = 0
position.reset_index(inplace=True)
pitchers["gameDate"] = pd.to_datetime(pitchers["gameDate"])
pitchers['season'] = pd.DatetimeIndex(pitchers['gameDate']).year
pitchers["days_between_game"] = pitchers.groupby(["name", "season"])["gameDate"].transform(lambda X: X.diff())
pitchers["days_between_game"] = pitchers["days_between_game"].apply(lambda x: x.days - 1)
pitchers[pitchers["days_between_game"] < 0] = 0
pitchers.reset_index(inplace=True)
```

```python
injury_pitchers = pitchers[pitchers.index.isin(test["idxmax"])][["name", "gameDate", "days_between_game"]]
injury_pitchers = injury_pitchers.merge(injury_data, on="name", how="left")
injury_pitchers[~injury_pitchers["injury"].isna()].sort_values("days_between_game", ascending=False)
injury_pitchers["before_injury"] = injury_pitchers.apply(lambda X: pd.to_datetime(X["date"]) - pd.Timedelta(days = X["days_between_game"] + 2), axis =1)
injury_pitchers["after_injury"] = injury_pitchers.apply(lambda X: pd.to_datetime(X["date"]) + pd.Timedelta(days = X["days_between_game"] + 2), axis =1)
pitcher_injuries = injury_pitchers[injury_pitchers["gameDate"].between(injury_pitchers["before_injury"], injury_pitchers["after_injury"])]
pitcher_injuries["injury"].unique()
```
