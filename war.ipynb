```python
df = pd.concat([position_minors, position_major]).sort_values(by=['name', 'gameDate'])
df
```

```python
df["gameDate"] = pd.to_datetime(df["gameDate"])
df['season'] = pd.DatetimeIndex(df['gameDate']).year
df["days_between_game"] = df.groupby(["name", "season"])["gameDate"].transform(lambda X: X.diff())
df["days_between_game"] = df["days_between_game"].apply(lambda x: x.days - 1)
df.reset_index(inplace=True)
```

```python
df[(df["name"] == "Brandon Nimmo") & (df["season"] == 2018)].sort_values("gameDate")[100:]
def is_callup(X):
    print(X)
    if (X["days_between_game"] > 7) & (df[X.index - 1]["league"] != X.league):
        return 1
    else:
        return 0
# df.transform(is_callup, axis=1)
```

```python
df.sort_values("days_between_game", ascending=False)[~df["league"].isin(["National League", "American League"])]
```

```python
days_missed = df.groupby("name").apply(lambda X: X[["days_between_game"]].sum() 
                            if any(X["league"] == "National League") or any(X["league"] == "American League") else 0).reset_index()
days_missed
```

```python
graph_df = position_war.merge(days_missed, left_on="Name", right_on="name", how="left")
```
