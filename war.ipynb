```python
position_minors = pd.read_csv("position_minors.csv")
position_major = pd.read_csv("position_majors.csv")
```

```python
df = pd.concat([position_minors, position_major]).sort_values(by=['name', 'gameDate'])
df
```

```python
df["gameDate"] = pd.to_datetime(df["gameDate"])
df['season'] = pd.DatetimeIndex(df['gameDate']).year
df["days_between_game"] = df.groupby(["name", "season"])["gameDate"].transform(lambda X: X.diff())
df["days_between_game"] = df["days_between_game"].apply(lambda x: x.days)
df.reset_index(inplace=True)
```
